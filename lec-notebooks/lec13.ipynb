```python
### PSTAT 8, Lecture 13

import numpy as np
import matplotlib.pyplot as plt
import sys
import scipy as sp
```

```python
# definition of our first function
def f(x):
    return (np.exp(x) - 2)
```

```python
x = 0
f(x)
```

```python
x = 1
f(x)
```

```python
f(2)
```

```python
f(10)
```

```python
# let's plot a line graph of f(x) from 0 to 3
# we need our x and y's first

n = 100
x = np.arange(n)*3.0/n
y = np.zeros(n) # initialize y to match the dimensions of x
```

```python
x
```

```python
y
```

```python
for i in range(n):
    y[i] = f(x[i])
    print(i,y[i], f(x[i]))
```

```python
# now that we have our x and y's, okay let's finally plot!
plt.figure(figsize=(9,8),dpi=80)
plt.ylim(-1,10)
plt.xlim(0,3)
plt.plot(x,y)

plt.xlabel('x', fontsize = 30)
plt.ylabel('f(x)', fontsize = 30)
print('f(x) = exp(x) - 2')
```
